```python
def interkuartil(data,x):
  q1 = (df[x]).quantile(0.25)
  q3 = (df[x]).quantile(0.75)
  iqr = q3-q1
  max = q3 + (1.5*iqr)
  min = q1 - (1.5*iqr)
  return max,min
```

```python
max,min = interkuartil(df,'max')
print('Nilai Maksimum: ',max, ' Nilai Minimum: ', min)
```

```python
def handle_outliers(df,x,max,min):
  lebihdari = (df[x]>max)
  kurangdari = (df[x]>min)
  df[x] = df[x].mask(lebihdari, max, axis=0)
  df[x] = df[x].mask(kurangdari, min, axis=0)
  return df
```

```python
max,min = interkuartil(df,'max')
df = handle_outliers(df,'max',max,min)
bp = df.boxplot(column=['max'])
```

```python
from sklearn.preprocessing import MinMaxScaler

normalisasi = MinMaxScaler()
deta = normalisasi.fit_transform(df)
col = [x,y]
df = pd.DataFrame(deta, columns=col)
```

```python
plot.scatter(df[x], df[y])
plot.xlabel(x)
plot.ylabel(y)
```

```python
def inisiasi_centroid(df,x,y,j):
  rnd.seed(10)
  centroids = {i+1:[rnd.choices(df[x]),rnd.choices(df[y])] for i in range(j)}
  return centroids
```

```python
j = 2
sentroid = inisiasi_centroid(df,x,y,j)
for i in sentroid:
  print(sentroid[i])

plot.scatter(df[x], df[y], color='green', alpha=0.5, edgecolor='black')
for i in sentroid:
  plot.scatter(sentroid[i][0],sentroid[i][1],color='red', s=400)
plot.show()
```

```python
def hitungjarak(data,x,y,centroid):
  for i in centroid.keys():
    df[str(i)] = np.sqrt((data[x]-centroid[i][0])**2 + (df[y]-centroid[i][1])**2)
  return df
```

```python
df2 = hitungjarak(df,x,y,sentroid)
df2
```

```python
petawarna = ['red','blue','green']
```

```python
def keanggotaan_data(df, centroid):
  df['index cluster'] = (df.loc[:, ['{}'.format(i) for i in centroid.keys()]].idxmin(axis=1)).astype('int')
  df['color'] = df['index cluster'].map(lambda x: petawarna[x])
  return df
```

```python
df2 = keanggotaan_data(df2,sentroid)
df2
```

```python
plot.scatter(df2[x], df2[y], color=df2['color'], alpha=0.5, edgecolor='black')
for i in sentroid:
  plot.scatter(sentroid[i][0],sentroid[i][1],color='yellow', s=400)
plot.show()
```

```python
def centroid_baru(df,centroid):
  for i in centroid.keys():
    centroid[i][0] = np.mean(df[df['index cluster'] == i][x])
    centroid[i][1] = np.mean(df[df['index cluster'] == i][y])
  return centroid
```

```python
def clusterisasi(df,x,y,k):
  centroid = inisiasi_centroid(df,x,y,k)
  df2 = copy.deepcopy(df)
  df2 = hitungjarak(df2,x,y,centroid)
  df2 = keanggotaan_data(df2,centroid)
  centroid = centroid_baru(df2,centroid)
  while (True):
    centroidlama = copy.deepcopy(centroid)
    df2 = hitungjarak(df2,x,y,centroid)
    df2 = keanggotaan_data(df2,centroid)
    centroid = centroid_baru(df2,centroid)
    if (centroidlama == centroid):
      break
  return (df2,centroid)
```

```python
k=2
df2,centroid = clusterisasi(df,x,y,k)

plot.scatter(df2[x], df2[y], color=df2['color'], alpha=0.5, edgecolor='black')
for i in centroid:
  plot.scatter(*centroid[i],color='blue', s=400)
plot.show()
```

```python
def sse(df,centroid):
  sum = 0
  for i in range(len(centroid)):
    j = i+1
    k = df[(df['index cluster'] == j)]['{}'.format(j)]**2
    sum += k.sum()
  return sum
```

```python
wcss = []
for i in range(2):
  df2, centroid = clusterisasi(df,x,y,i+1)
  wcss.append(sse(df2, centroid))
  print('K=',i+1,'  Distorsi: ',wcss[i])
```
